```python
import pandas as pd
from datetime import datetime
```

```python
# 파일 불러오기
import glob

path = r'./data' # use your path
all_files = glob.glob(path + "/210901/13*.csv")

count = len(all_files)
print("Reading", count, "Files")

li = []

for i, filename in enumerate(all_files):
    df = pd.read_csv(filename, index_col=None, header=0)
    
    filedatelist = filename.split("/")
    filedate = filedatelist[2] + filedatelist[3][:-4]
#     print(datetime.strptime(filedate, "%y%m%d%H%M%S"))
#     print(filedate)
    
    df["time"] = datetime.strptime(filedate, "%y%m%d%H%M%S")
    
    li.append(df)
    
    if (i + 1) % 100 == 0:
        print(f"Read {i + 1}/{count}")

print("Read Complete")

df = pd.concat(li, axis=0, ignore_index=True)
print("Concat Complete")
```
